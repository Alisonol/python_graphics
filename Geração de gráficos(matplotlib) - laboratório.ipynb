# Gráfico de barras

Utilizado para representar dados categóricas em formato de barras retangulares

```python
x1 = [2, 4, 6, 8, 10]
y1 = [6, 7, 8, 2, 4]
```

```python
plt.bar(x1, y1, label = 'Barras', color = 'green')
plt.legend()
plt.show()
```

```python
x2 = [1, 3, 5, 7, 9]
y2 = [7, 8, 2, 4, 2]
```

```python
plt.bar(x1, y1, label = 'Listas1', color = 'blue')
plt.bar(x2, y2, label = 'Listas2', color = 'red')
plt.legend()
plt.show()
```

```python
import random

idades = random.randint(40, 120)
print(idades)
```

```python
#Geração aleatória de números com um range informado

import random

idades = []

for i in range(0,25):
    x = random.randint(1,120)
    idades.append(x)

print(idades)
```

```python
# List comprehension para obter os índices respectivos dos números 

ids = [x for x in range(len(idades))]
```

```python
plt.bar(ids, idades, label = 'Gráfico de idades', color = 'pink')
plt.legend()
plt.show()
```

# Histograma de barras

```python
bins = []

for i in range(0,131, 5):
    bins.append(i)
print(bins)
```

```python
plt.hist(idades, bins, histtype = 'bar', rwidth = 0.9)
plt.show()
```

```python
plt.hist(idades, bins, histtype = 'stepfilled', rwidth = 0.9)
plt.show()
```

# Gráfico de pizza

Representa a composição que representa a composição de uma variável categórica em relação ao todo. Usados para mostrar a distribuição de dados em relação ao todo e para destacar as proporções relativas de diferentes categorias.

```python
fatias = [7, 2, 2, 13]
atividades = ['dormir', 'comer', 'passear', 'trabalhar']
cores = ['olive', 'lime', 'violet', 'royalblue']
```

```python
fig, ax = plt.subplots(figsize=(5, 6), subplot_kw=dict(aspect="equal"))
plt.pie(fatias, labels = atividades, colors = cores, startangle = 90, shadow = True, explode = [0, 0.2, 0, 0], autopct='%1.1f%%')
plt.legend(atividades, loc='upper right')
plt.axis('equal')
plt.show()
```
