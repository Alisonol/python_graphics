# Stastitical Data Visualization com Seaborn

```python
# Imports 
import random 
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import seaborn as sea
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# Carregando um dos datasets que vÃªm com o seaborn

dados = sea.load_dataset("tips")
dados.head()
```

```python
# O mÃ©todo joinplot cria plot de 2 variÃ¡vveis com grÃ¡ficos bivariados e univariados

sea.jointplot(data = dados, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
# O mÃ©todo lmplot() cria plot com dados e modelos de regressÃ£o

sea.lmplot(data = dados, x = 'total_bill', y = 'tip', col = 'smoker')
```
