```python
# O pylab combina funcionalidades do pyplot com funcionalidade do Numpy

from pylab import *
%matplotlib inline
```

# Gráfico de linha

Utilizado quando desejamos avaliar a tendência, comportamento de uma variável ao longo do tempo. Em especial, buscamos checar a flutuação desta variável no período proposto

```python
# Gráfico de linha

# Dados

x = linspace(0, 5, 10) # Números com espaçamento uniforme
y = x ** 2


# Cria a figura
fig = plt.figure()

# Define a escala dos eixos 
axes = fig.add_axes([0, 0, 0.8, 0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e título

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de linha')
```

```python
# Gráfico de linha com 2 figuras

# Dados
x = linspace(0, 5, 10) # Números com espaçamento uniforme
y = x ** 2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos 
axes1 = fig.add_axes([0, 0, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.1, 0.4, 0.4, 0.3]) # eixos da figura secundária

# Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura principal')

# Figura secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura secundária')
```

```python
# Gráfico de linha em paralelo

# Dados
x = linspace(0, 5, 10) # Números com espaçamento uniforme
y = x ** 2

# Cria a figura
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para cirar cada plot
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
# Ajusta o layout
fig.tight_layout()
```

```python
# Gráficos de linha com diferentes escalas

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Cria o plot 1
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padrão")

# Cria o plot 2
axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logarítmica (y)")
```

```python
# Grid

# Dados
x = linspace(0, 5, 10)
y = x ** 2

# Cria os subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Grid Padrão
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)
axes[0].set_title("Grid Padrão")

# Grid customizado
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color = 'k', alpha = 0.2, linestyle = 'dashed', linewidth = 0.8)
# Dentro do alpha, trabalhamos com a opacidade, quanto maior o número, menos opaco é e o inverso é verdadeiro
axes[1].set_title("Grid Customizado")
```

```python
# Diferentes estilos de plots

# Dados

xx = linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

# Subplots
fig, axes = plt.subplots(1, 4, figsize = (12, 3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = 'k')
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n ** 2, lw = 2, color = 'b')
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n ** 2, align = 'center', width = 0.5, alpha = 0.5, color = 'm' )
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x ** 2, x ** 3, alpha = 0.5, color = 'green')
axes[3].set_title("fill_between")
```

```python
# Histogramas

# Dados
n = np.random.randn(100000)

# Cria os subplots 
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

# Plot 1 
axes[0].hist(n)
axes[0].set_title("Hitograma Padrão")
axes[0].set_xlim((min(n)), max(n))

# Plot 2

axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)))
```
